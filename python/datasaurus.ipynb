# Seccion utilizando mi punto principal, en mi caso la variable AWAY

```python
import pandas as pd
# Cargar el dataset
data = pd.read_csv('DatasaurusDozen.tsv', sep='\t')

# Filtrar el dataset para solo utilizar 'away'
data_away = data[data['dataset'] == 'away']

# Ver las primeras filas del nuevo dataset
data_away.head()
```

# Calculo de promedio, media, varianza, desviacion estandar

```python
datasets_grouped = data_away.groupby('dataset')
datasets_grouped.agg(['count', 'mean', 'var', 'std'])
```

# Calculo de correlaci贸n

```python
corr = data_away.groupby('dataset')[['x','y']].corr().iloc[0::2,-1]
corr
```

# Calculo e covarianza 

```python
cov_matrix = data_away.groupby('dataset').cov()
cov_matrix
```

# Visualizaci贸n general

```python
grid_scatterplots = sns.FacetGrid(data_away, col="dataset", hue="dataset", col_wrap=4)
grid_scatterplots.map_dataframe(sns.scatterplot, x="x", y="y")
```

# Regresi贸n Lineal

```python
sns.lmplot(data=data_away, x="x", y="y", col="dataset", hue="dataset", line_kws={'color': '#003f5c'}, col_wrap=4, ci=None, height=3)
```

```python
from sklearn.linear_model import LinearRegression


# Filtrar el dataset para solo utilizar 'away'
data_away = data[data['dataset'] == 'away']

model = LinearRegression()

X = data_away[['x']]
y = data_away['y']
model.fit(X, y)

intercept = model.intercept_
slope = model.coef_[0]

print("Intercept:", intercept)
print("Slope:", slope)
```
